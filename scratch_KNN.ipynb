```python
# -*- coding: utf-8 -*-
"""
Created on Wed Jul  8 18:37:54 2020

@author: Priyanshu
"""
import numpy as np 
import pandas as pd 
from sortedcontainers import SortedList
from datetime import datetime
print("reading & transforming data to matrix ") 
dataset=pd.read_csv('Social_Network_Ads.csv') 
x=dataset.iloc[:,0:2].values
y=dataset.iloc[:,2:].values

# spliting test train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y.ravel(), test_size=0.25,random_state=0 )

# feature scaling
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)


class knn(object):
    def __init__(self,k):
        self.k=k
    
    def fit(self,a,b):
        self.a=a
        self.b=b
    
    def predict(self,a):
        ypred=np.zeros(len(a))
        
        for i,val in enumerate(a):
           sl=SortedList()
           for j,xt in enumerate(self.a):
               diff=val-xt
               d=diff.dot(diff)
               if len(sl)<self.k :
                   sl.add((d,self.b[j]))
               else :
                   if d<sl[-1][0] :
                       del sl[-1]
                       sl.add((d,self.b[j]))
           votes={}
           for _,v in sl :
               votes[v]=votes.get(v,0) + 1 
               max_votes=0
               max_votes_class=-1
               for v,count in votes.items():
                   if count> max_votes:
                       max_votes=count
                       max_votes_class=v
                       
           ypred[i]=max_votes_class
        return ypred
    
    def score(self,a,b) :
        p=self.predict(a)
        from sklearn.metrics import confusion_matrix
        cm=confusion_matrix(b,p)

        return np.mean(p==b),cm
    
for k in (1,2,3,4,5) :
    print("K===>",k)
    KNN=knn(k)
    t0=datetime.now()
    KNN.fit(x_train,y_train)
    print("train time",(datetime.now()-t0))
    
    t0=datetime.now()
    print("train accracy",KNN.score(x_train,y_train ))
    print("time taken",(datetime.now()-t0))
    
    t0=datetime.now()
    print("train accuracy",KNN.score(x_test,y_test ))
    print("time taken ",(datetime.now()-t0))
  
```
